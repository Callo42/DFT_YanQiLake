```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

# load data
energy, dos, idos = np.loadtxt('../src/NaCl/dos/NaCl_dos.dat', unpack=True)

# make plot
plt.figure(figsize = (12, 6))
plt.plot(energy, dos, linewidth=0.75, color='red')
plt.yticks([])
plt.xlabel('Energy (eV)')
plt.ylabel('DOS')
plt.axvline(x=-0.820, linewidth=0.5, color='k', linestyle=(0, (8, 10)))
plt.xlim(-3, 18)
plt.ylim(0, )
plt.fill_between(energy, 0, dos, where=(energy < -0.820), facecolor='red', alpha=0.25)
plt.text(-0.9, 1.7, 'Fermi energy', rotation=90)
plt.show()
```
