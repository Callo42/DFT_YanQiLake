```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()
x, y = np.loadtxt('../src/Convergence/etot_vs_kpoint.dat', delimiter=' ', unpack=True)
ax.plot(x, y, "o-", markersize=5, label='E total -- kpoints')
ax.set_xlabel('sampled kpoints')
ax.set_ylabel('Etot (Ry)')
ax.legend()
ax.set(
    xlim=(2,16),
    # ylim=(0,1.3),
    xticks=range(2,17),
    # yticks=np.arange(15)/10
    )
plt.show()
```
